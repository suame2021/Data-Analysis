### Analysis is done on the data from year 1990-2020


```python
df = df[df['year']>=1990]
```

### Checking null values present in the particular column 

```python
df.isnull().sum()
```

```python
df.head(45)
```

*Upon inspecting the dataset, it can be observed that there are a total of 9 null values in bio_fuel electricty and biofuel_elec_per_capita. Furthermore, it is noteworthy that it is because the production of energy from biofuel was started in the 2000 hence there is no data present.*
*Therefore, it would be appropriate to update the "biofuel_electricity" and "biofuel_elec_per_capita" column by replacing the null values with 0, as there is no data present in accordance with these fields.*

```python
df["biofuel_elec_per_capita"]=df['biofuel_elec_per_capita'].fillna(0)
df["biofuel_electricity"]=df['biofuel_electricity'].fillna(0)

df.isnull().sum()
```

### Reseting the index by reset_index() func

```python
df = df.reset_index(drop='true')
```

```python
df.tail()
```

```python
df.isnull().sum()
```

**As we can observe that there ia a NAN value in population column for year 2020, it would be appropriate to update the value from the Indian Census Data where we can get the population of India for the year 2020**

```python
population_2020=1396387127
df['population'].fillna(population_2020, inplace=True)
```
